Plots of Infectious Disease Simulations

```python
import matplotlib.pyplot as plt
import numpy as np
```

Observing number of days the disease runs its course vs vaccination

```python
data1 = np.loadtxt("vac_vs_steps1.out")
data2 = np.loadtxt("vac_vs_steps2.out")
data3 = np.loadtxt("vac_vs_steps3.out")
data4 = np.loadtxt("vac_vs_steps4.out")
plt.plot(data1[:,0],data1[:,1],'o', label = "Transmission Probability = 20%")
plt.plot(data2[:,0],data2[:,1],'o', label = "Transmission Probability = 40%")
plt.plot(data3[:,0],data3[:,1],'o', label = "Transmission Probability = 60%")
plt.plot(data4[:,0],data4[:,1],'o', label = "Transmission Probability = 80%")
plt.title("Days disease runs its course vs proportion vaccinated")
plt.xlabel("Proportion of population vaccinated"); plt.ylabel("Days the disease runs its course")
plt.legend()
plt.savefig("vac_vs_steps1.png")
# Population size 10000
```

```python
data = np.loadtxt("herdimmunity_data.out")
plt.plot(data[:,0],data[:,1],'o')
plt.title("Vaccination proportion required for herd immunity")
plt.xlabel("Probability of transmission"); plt.ylabel("Proportion of population vaccinated")
plt.savefig("herd_immunity_plot.png")
# Population size 10000
```

```python
mutation_data = np.loadtxt("vac_vs_mutationrate.out")
plt.plot(mutation_data[:,0],mutation_data[:,1],'o')
plt.title("Vaccination proportion required to prevent mutation")
plt.xlabel("Transmissions required for mutation"); plt.ylabel("Proportion of population vaccinated")
plt.savefig("prevent_mutation.png")
```
